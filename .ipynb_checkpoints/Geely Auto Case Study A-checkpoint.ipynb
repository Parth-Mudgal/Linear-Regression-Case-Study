## Step 2: Data Cleaning & Preparation

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
# droppping car_ID column
df = df.drop(columns = 'car_ID')
df.head()
```

```python
df.symboling.value_counts()
```

```python
# we need to convert symboling column to good or bad as 0 and 1 respectively
df['symboling'] = df['symboling'].apply(lambda x: 1 if x >= 3 else 0)
df.symboling.value_counts()
```

```python
# lets fix the CarName column now
df['CarName'] = df['CarName'].apply(lambda x: x.split(' ')[0])
df['CarName'].value_counts()
```

```python
df.fueltype.value_counts()
```

```python
# mapping 1 and 0 for gas and disel
df['fueltype'] = df['fueltype'].map({'gas': 1, 'diesel': 0})
df.fueltype.value_counts()
```

```python
df.aspiration.value_counts()
```

```python
# mapping 1 and 0 for aspiration
df['aspiration'] = df['aspiration'].map({'std': 1, 'turbo': 0})
df.aspiration.value_counts()
```

```python
df.doornumber.value_counts()
```

```python
# mapping 1 and 0 for doornumber as 1 for 4 and 0 for two
df['doornumber'] = df['doornumber'].map({'four': 1, 'two': 0})
df.doornumber.value_counts()
```

```python
df.enginelocation.value_counts()
```

```python
# mapping 1 and 0 for enginelocation as 1 for front and 0 for rear
df['enginelocation'] = df['enginelocation'].map({'front': 1, 'rear': 0})
df.enginelocation.value_counts()
```

```python
df.head()
```

 Now we have to deal with dummy variables that has more than 2 levels

```python
# let's deal with carbody first
df.carbody.value_counts()
```

```python
# getting dummy variables from carbody
car_body_types = pd.get_dummies(df['carbody']).astype('int')
car_body_types.head()
```

```python
# Let's drop the first column from car_body_types df using 'drop_first = True'
car_body_types = pd.get_dummies(df['carbody'], drop_first = True).astype('int')

# Add the results to the original dataframe

df = pd.concat([df, car_body_types], axis = 1)

# Drop 'carbody' as we have created the dummies for it

df.drop(['carbody'], axis = 1, inplace = True)

# let's check the df
df.head()
```

```python
# let's deal with drivewheel
df.drivewheel.value_counts()
```

```python
# Let's drop the first column from drivewheeltypes df using 'drop_first = True'
drivewheeltypes = pd.get_dummies(df['drivewheel'], drop_first = True).astype('int')

# Add the results to the original dataframe

df = pd.concat([df, drivewheeltypes], axis = 1)

# Drop 'drivewheel' as we have created the dummies for it

df.drop(['drivewheel'], axis = 1, inplace = True)

# let's check the df
df.head()
```

```python
# let's deal with enginetype
df.enginetype.value_counts()
```

```python
# Let's drop the first column from enginetypes df using 'drop_first = True'
enginetypes = pd.get_dummies(df['enginetype'], drop_first = True).astype('int')

# Add the results to the original dataframe

df = pd.concat([df, enginetypes], axis = 1)

# Drop 'enginetype' as we have created the dummies for it

df.drop(['enginetype'], axis = 1, inplace = True)

# let's check the df
df.head()
```

```python
# let's deal with cylindernumber
df.cylindernumber.value_counts()
```

```python
# Let's drop the first column from cylindernumbers df using 'drop_first = True'
cylindernumbers = pd.get_dummies(df['cylindernumber'], drop_first = True).astype('int')

# Add the results to the original dataframe

df = pd.concat([df, cylindernumbers], axis = 1)

# Drop 'cylindernumber' as we have created the dummies for it

df.drop(['cylindernumber'], axis = 1, inplace = True)

# let's check the df
df.head()
```

```python
# let's deal with fuelsystem
df.fuelsystem.value_counts()
```

```python
# Let's drop the first column from fuelsystems df using 'drop_first = True'
fuelsystems = pd.get_dummies(df['fuelsystem'], drop_first = True).astype('int')

# Add the results to the original dataframe

df = pd.concat([df, fuelsystems], axis = 1)

# Drop 'fuelsystem' as we have created the dummies for it

df.drop(['fuelsystem'], axis = 1, inplace = True)

# let's check the df
df.head()
```

```python
# checking Caranme column
df.CarName.value_counts()
```

 We have now dealt with all categorical columns and dummy variables

## Step 5: Building a linear model using RFE

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

num_features_to_select = 10

rfe = RFE(estimator=lm, n_features_to_select=num_features_to_select) # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```
